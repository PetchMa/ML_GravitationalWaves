```python
import pandas as pd 
import os
import torch
import torchvision
from torch.utils import data as D
import numpy as np
import pandas as pd
import gc
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

import os
import glob
import os.path as osp
from PIL import Image

import matplotlib.pyplot as plt

data_dir = '../input/g2net-gravitational-wave-detection'

labels = pd.read_csv('../input/g2net-gravitational-wave-detection/training_labels.csv')
labels
```

```python
import os

li = ['0','1','2','3','4','5','6','7','a','b','c','d','e','f']
count = 0
train_data = []
for i in range(1):
    for j in range(len(li)):
        for k in range(len(li)):
            for first_dir in os.listdir(data_dir+'/train/'+str(li[i])+'/'+str(li[j])+'/'+str(li[k]) ):
                train_data.append(np.load(data_dir+'/train/'+str(li[i])+'/'+str(li[j])+'/'+str(li[k])+ '/'+first_dir)) 
                
                
train_data = np.array(train_data)
train_data = np.swapaxes(train_data, 1, 2)
print(train_data.shape)
```

```python
train_labels = np.array(labels['target'].to_list())
train_labels = train_labels[:train_data.shape[0]]
print(train_labels.shape, train_data.shape)
print(train_labels.shape)

X_train, X_test = train_test_split(train_data,  test_size=0.33, random_state=42)
y_train, y_test = train_test_split(train_labels,  test_size=0.33, random_state=42)
del train_data
gc.collect()
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train =(X_train + abs(X_train.min()))
X_train = X_train/X_train.max()
X_test = (X_test + abs(X_test.min()))
X_test = X_test/X_test.max()
print(X_train.max(), X_train.min())
```

```python
import numpy
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
# create the model
embedding_vecor_length = 3
length_seq = 4096
model = Sequential()
# model.add(input())
# model.add(Embedding(length_seq, embedding_vecor_length, input_length=length_seq))
model.add(LSTM(300))

model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])


model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3, batch_size=256)
```
